```python
FUNCTIONS

Q1. Write a function greet() that prints "Hello World".

def greet():
    print("Hello World")

greet()   # Output: Hello World


Q2. Write a function add(a, b) that returns the sum of two numbers.

def add(a, b):
    return a + b

print(add(5, 3))   # Output: 8


Q3. Write a function with default argument: greet(name="Guest").

def greet(name="Guest"):
    print("Hello", name)

greet()           # Output: Hello Guest
greet("Alice")    # Output: Hello Alice


Q4. Write a function even_or_odd(n) that prints whether a number is even or odd.

def even_or_odd(n):
    if n % 2 == 0:
        print("Even")
    else:
        print("Odd")

even_or_odd(7)   # Output: Odd


Q5. Write a function that takes variable number of arguments and returns their sum.

def sum_all(*args):
    return sum(args)

print(sum_all(1, 2, 3, 4))   # Output: 10


Q6. Write a function that returns multiple values (sum and product).

def sum_and_product(a, b):
    return a+b, a*b

s, p = sum_and_product(3, 4)
print(s, p)   # Output: 7 12


Q7. Write a recursive function to find factorial of a number.

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(5))   # Output: 120


Q8. Write a lambda function to square a number.

square = lambda x: x**2
print(square(6))   # Output: 36


Q9. Use a function inside a list comprehension: square numbers 1 to 5.

def square(x):
    return x**2

squares = [square(x) for x in range(1,6)]
print(squares)   # Output: [1, 4, 9, 16, 25]


Q10. Write a function to check if a string is palindrome.

def is_palindrome(s):
    return s == s[::-1]

print(is_palindrome("madam"))   # Output: True
print(is_palindrome("hello"))   # Output: False

MODULES

Q1. Import math module and print square root of 16.

import math
print(math.sqrt(16))   # Output: 4.0


Q2. Import random module and print a random integer between 1 and 10.

import random
print(random.randint(1, 10))   # Output: random number 1-10


Q3. Import only pi from math module and print it.

from math import pi
print(pi)   # Output: 3.141592653589793


Q4. Use os module to get current working directory.

import os
print(os.getcwd())   # Output: current directory path


Q5. Use datetime module to print current date and time.

import datetime
print(datetime.datetime.now())   # Output: current date and time


Q6. Use random.choice() to select a random element from a list.

import random
lst = [10, 20, 30, 40]
print(random.choice(lst))   # Output: random element from list


Q7. Use math module to calculate factorial of 5.

import math
print(math.factorial(5))   # Output: 120


Q8. Create your own module my_module.py with a function greet() and import it.

# my_module.py
def greet(name):
    print("Hello", name)

# main.py
from my_module import greet
greet("Alice")   # Output: Hello Alice


Q9. Use dir() to list all attributes of math module.

import math
print(dir(math))
# Output: list of all functions/constants in math module


Q10. Use help() to get documentation of random.randint.

import random
help(random.randint)
# Output: shows how randint() works
```
